1. Write a python program which searches all the product under a particular product from www.amazon.in. The 
product to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for 
guitars.

```python
import requests
from bs4 import BeautifulSoup

def search_amazon_product(product_name):
    base_url = 'https://www.amazon.in/s?k='
    product_name = product_name.replace(' ', '+')
    url = base_url + product_name
    
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    products = soup.find_all('div', {'data-component-type': 's-search-result'})
    
    for product in products:
        product_name = product.find('span', {'class': 'a-size-base-plus a-color-base a-text-normal'}).text.strip()
        print(product_name)

if __name__ == "__main__":
    product_name = "guitar"
    search_amazon_product(product_name)
```

2. In the above question, now scrape the following details of each product listed in first 3 pages of your search 
results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then 
scrape all the products available under that product name. Details to be scraped are: "Brand 
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and 
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def scrape_amazon_products(product_name, num_pages):
    base_url = f'https://www.amazon.in/s?k={product_name.replace(" ", "+")}'
    data = []

    for page in range(1, num_pages+1):
        url = f"{base_url}&page={page}"
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        products = soup.find_all('div', {'data-component-type': 's-search-result'})
        
        for product in products:
            name = product.find('span', {'class': 'a-size-base-plus a-color-base a-text-normal'})
            brand = product.find('span', {'class': 'a-size-base-plus a-color-base'})
            price = product.find('span', {'class': 'a-price-whole'})
            return_exchange = product.find('span', {'class': 'a-icon-alt'})
            delivery = product.find('span', {'class': 'a-text-bold'})
            availability = product.find('span', {'class': 'a-size-base'})
            url = 'https://www.amazon.in' + product.find('a', {'class': 'a-link-normal'})['href'] if product.find('a', {'class': 'a-link-normal'}) else '-'
            
            data.append([brand.text.strip() if brand else '-', 
                         name.text.strip() if name else '-', 
                         price.text.strip() if price else '-', 
                         return_exchange.text.strip() if return_exchange else '-', 
                         delivery.text.strip() if delivery else '-', 
                         availability.text.strip() if availability else '-', 
                         url])
            
    return data
```

```python
def save_to_csv(data, filename):
    df = pd.DataFrame(data, columns=["Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Product URL"])
    df.to_csv(filename, index=False)
    print("Data saved to", filename)
```

```python
if __name__ == "__main__":
    product_name = input("Enter the product name: ")
    num_pages = int(input("Enter the number of pages to scrape: "))
    
    product_data = scrape_amazon_products(product_name, num_pages)
    save_to_csv(product_data, "amazon_products.csv")
    
```

```python
import requests
from bs4 import BeautifulSoup
import re

def scrape_google_images(keyword, num_images):
    # Construct the search URL
    search_url = f"https://www.google.com/search?q={keyword}&tbm=isch"
    
    # Send a GET request to Google Images
    response = requests.get(search_url)
    
    # Parse the HTML content
    soup = BeautifulSoup(response.content, "html.parser")
    
    # Find all image elements
    image_elements = soup.find_all("img", class_="rg_i Q4LuWd")
    
    # Extract image URLs
    image_urls = []
    for img in image_elements[:num_images]:
        try:
            # Extract image URL from data-src attribute
            image_url = img["data-src"]
            image_urls.append(image_url)
        except KeyError:
            pass
    
    return image_urls

# Define keywords and number of images to scrape
keywords = ['fruits', 'cars', 'Machine Learning', 'Guitar', 'Cakes']
num_images_per_keyword = 10

# Scrape images for each keyword
for keyword in keywords:
    print(f"Scraping images for keyword: {keyword}")
    image_urls = scrape_google_images(keyword, num_images_per_keyword)
    
    # Print image URLs
    print(f"Images for keyword '{keyword}':")
    for url in image_urls:
        print(url)
    print("\n")
```

4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com
and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand 
Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, 
“Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the 
details is missing then replace it by “- “. Save your results in a dataframe and CSV.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_flipkart_smartphones(search_query):
    url = f"https://www.flipkart.com/search?q={search_query}&page=1"
    headers = {"User-Agent": "Mozilla/5.0"}
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    smartphone_data = []
    
    products = soup.find_all('div', {'class': '_1AtVbE'})
    for product in products:
        try:
            brand = product.find('div', {'class': '_4rR01T'}).text.strip()
        except AttributeError:
            brand = '-'
        
        try:
            name = product.find('a', {'class': 'IRpwTa'}).text.strip()
        except AttributeError:
            name = '-'
        
        try:
            color = product.find('a', {'class': 'IRpwTa'}).next_sibling.next_sibling.text.strip()
        except AttributeError:
            color = '-'
        
        details = product.find_all('li', {'class': 'rgWa7D'})
        details_dict = {}
        for detail in details:
            key = detail.find('div', {'class': 'rgWa7D'}).text.strip()
            value = detail.find('li').text.strip()
            details_dict[key] = value
        
        ram = details_dict.get('RAM', '-')
        rom = details_dict.get('Storage (ROM)', '-')
        primary_camera = details_dict.get('Primary Camera', '-')
        secondary_camera = details_dict.get('Secondary Camera', '-')
        display_size = details_dict.get('Display Size', '-')
        battery_capacity = details_dict.get('Battery Capacity', '-')
        
        try:
            price = product.find('div', {'class': '_30jeq3 _1_WHN1'}).text.strip()
        except AttributeError:
            price = '-'
        
        try:
            product_url = 'https://www.flipkart.com' + product.find('a', {'class': 'IRpwTa'})['href']
        except AttributeError:
            product_url = '-'
        
        smartphone_data.append({
            'Brand Name': brand,
            'Smartphone Name': name,
            'Colour': color,
            'RAM': ram,
            'Storage(ROM)': rom,
            'Primary Camera': primary_camera,
            'Secondary Camera': secondary_camera,
            'Display Size': display_size,
            'Battery Capacity': battery_capacity,
            'Price': price,
            'Product URL': product_url
        })
    
    return smartphone_data

def save_to_csv(data, filename):
    df = pd.DataFrame(data)
    df.to_csv(filename, index=False)
    print("Data saved to", filename)

if __name__ == "__main__":
    search_query = input("Enter the smartphone you want to search for: ")
    data = scrape_flipkart_smartphones(search_query)
    save_to_csv(data, "flipkart_smartphones.csv")
```

5. Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps.

```python
import requests
from bs4 import BeautifulSoup

def get_coordinates(city):
    try:
        # Format the city name for the URL
        formatted_city = city.replace(" ", "+")

        # Send a GET request to Google Maps with the formatted city name
        url = f"https://www.google.com/maps/search/{formatted_city}"
        print("URL:", url)  # Debug print
        response = requests.get(url)
        print("Response status code:", response.status_code)  # Debug print
        response.raise_for_status()

        # Parse the HTML response using BeautifulSoup
        soup = BeautifulSoup(response.text, "html.parser")
        print("HTML:", soup)  # Debug print

        # Find the element containing the coordinates
        coordinates_element = soup.find("meta", itemprop="image")
        print("Coordinates element:", coordinates_element)  # Debug print

        # Extract the coordinates from the element's content attribute
        if coordinates_element:
            coordinates = coordinates_element["content"].split(";")[1].strip().split(",")
            latitude, longitude = float(coordinates[0]), float(coordinates[1])
            return latitude, longitude
        else:
            print("Coordinates element not found.")
            return None
    except Exception as e:
        print("Error occurred:", e)
        
        return None

if __name__ == "__main__":
    city = input("Enter a city name: ")
    coordinates = get_coordinates(city)
    if coordinates:
        latitude, longitude = coordinates
        print(f"The coordinates of {city} are: Latitude: {latitude}, Longitude: {longitude}")
    else:
        print("No coordinates found.")
```

7. Write a python program to scrape the details for all billionaires from www.forbes.com. Details to be scrapped: 
“Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”, “Industry”.

```python
import requests
from bs4 import BeautifulSoup

def scrape_billionaires_table(url):
    try:
        # Send a GET request to the provided URL
        response = requests.get(url)
        response.raise_for_status()

        # Create a BeautifulSoup object to parse the HTML content
        soup = BeautifulSoup(response.content, "html.parser")

        # Find the correct table that contains the billionaire details
        table = soup.find("table", class_="table")

        if table:
            # Find all rows in the table
            rows = table.find_all("tr")

            # Iterate over each row and extract the required details
            for row in rows[1:]:  # Skip the header row
                # Find all columns in the row
                columns = row.find_all("td")
                
                # Extract the required details from the columns
                rank = columns[0].text.strip()
                name = columns[1].text.strip()
                net_worth = columns[2].text.strip()
                age = columns[3].text.strip()
                citizenship = columns[4].text.strip()
                source = columns[5].text.strip()
                industry = columns[6].text.strip()
                
                # Print the extracted details
                print("Rank:", rank)
                print("Name:", name)
                print("Net Worth:", net_worth)
                print("Age:", age)
                print("Citizenship:", citizenship)
                print("Source:", source)
                print("Industry:", industry)
                print()
        else:
            print("Table not found on the page.")
    except Exception as e:
        print("Error:", e)

if __name__ == "__main__":
    url = "https://www.forbes.com/billionaires/"
    scrape_billionaires_table(url)
```

8. Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted 
from any YouTube Video.

```python
from selenium import webdriver
from bs4 import BeautifulSoup
from datetime import datetime
import time

def extract_comments_selenium(video_url):
    comments = []

    # Set up the Chrome webdriver
    options = webdriver.ChromeOptions()
    options.add_argument("--headless")  # Run Chrome in headless mode, no GUI needed
    options.add_argument("--disable-gpu")
    options.add_argument("--no-sandbox")
    driver = webdriver.Chrome(options=options)  # Make sure chromedriver is in PATH or specify the path

    # Open the YouTube video URL
    print("Opening YouTube video page...")
    driver.get(video_url)
    print("YouTube video page opened.")

    # Scroll down to load comments
    print("Scrolling to load comments...")
    for _ in range(5):  # Adjust the range according to the number of times you want to scroll
        driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
        time.sleep(3)  # Wait for comments to load

    # Extract comments using BeautifulSoup
    print("Extracting comments...")
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    comment_elements = soup.find_all('yt-formatted-string', class_='style-scope ytd-comment-renderer', id='content-text')
    print(f"Found {len(comment_elements)} comment elements.")
    for comment_element in comment_elements:
        comment_text = comment_element.get_text()
        comment_upvotes = comment_element.find_previous('span', class_='style-scope ytd-comment-action-buttons-renderer')
        if comment_upvotes:
            comment_upvotes = comment_upvotes.get_text()
        else:
            comment_upvotes = '0'
        comment_time_element = comment_element.find_previous('a', class_='yt-simple-endpoint', id='published-time-text')
        if comment_time_element:
            comment_time = comment_time_element.get_text()
            comment_time = datetime.strptime(comment_time, '%b %d, %Y')
        else:
            comment_time = None
        comments.append({'text': comment_text, 'upvotes': comment_upvotes, 'time': comment_time})

    # Close the webdriver
    driver.quit()
    print("WebDriver closed.")

    return comments

# Specify the YouTube video URL
video_url = 'https://youtu.be/sPPFDBUJzA0?si=6rkx73z0smbDV5pi'

# Extract comments using Selenium
comments_selenium = extract_comments_selenium(video_url)

# Display the extracted comments
print("Displaying extracted comments:")
for i, comment in enumerate(comments_selenium):
    print(f"Comment {i+1}:")
    print(f"Text: {comment['text']}")
    print(f"Upvotes: {comment['upvotes']}")
    print(f"Time: {comment['time']}")
    print("\n")
```

9. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in 
“London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall 
reviews, privates from price, dorms from price, facilities and property description.

```python
import requests
from bs4 import BeautifulSoup

def scrape_hostels_in_london():
    url = "https://www.hostelworld.com/hostels/London"
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "html.parser")
        hostels = soup.find_all("div", class_="fabresult")
        hostels_data = []
        for hostel in hostels:
            name = hostel.find("h2", class_="fabresult-title").text.strip()
            distance = hostel.find("span", class_="distance").text.strip()
            ratings = hostel.find("div", class_="rating").text.strip()
            total_reviews = hostel.find("div", class_="reviews").text.strip()
            overall_reviews = hostel.find("div", class_="overall").text.strip()
            privates_price = hostel.find("div", class_="price-col").find("div", class_="price").text.strip()
            dorms_price = hostel.find_all("div", class_="price-col")[1].find("div", class_="price").text.strip()
            facilities = hostel.find("div", class_="facilities").text.strip()
            description = hostel.find("div", class_="description").text.strip()
            hostel_data = {
                "Name": name,
                "Distance from City Centre": distance,
                "Ratings": ratings,
                "Total Reviews": total_reviews,
                "Overall Reviews": overall_reviews,
                "Privates from Price": privates_price,
                "Dorms from Price": dorms_price,
                "Facilities": facilities,
                "Property Description": description
            }
            hostels_data.append(hostel_data)
        return hostels_data
    else:
        print("Failed to retrieve data from the website.")
        return None

# Scrape data for all available hostels in London
hostels_data = scrape_hostels_in_london()

# Print the scraped data
if hostels_data:
    for index, hostel_data in enumerate(hostels_data, start=1):
        print(f"Hostel {index}:")
        for key, value in hostel_data.items():
            print(f"{key}: {value}")
        print()
else:
    print("No data scraped.")
```
